```python
x=np.array([-7.,-4.,-1.,2.,5.,8.,11.,14.])
y=np.array([3.,6.,9.,12.,15.,18.,21.,24.])

#Since x contains your feature values, it should have the shape (num_samples, num_features)
#(n_samples,1)  ->   (8,1)
x = np.array([-7., -4., -1., 2., 5., 8., 11., 14.]).reshape(-1, 1)

tf.random.set_seed(42)
model=tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])
#model.summary()
model.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.SGD(),metrics=["mae"])
model.fit(x,y,epochs=5)
```

```python
model.summary()
```

Reshape: x.reshape(-1, 1) converts the 1D array into a 2D array with 8 samples and 1 feature.

# Model Training Process in TensorFlow/Keras

When you call `model.fit(x, y, epochs=5)`, the model goes through the following steps for each epoch:

## 1. Forward Pass
- **Input Data**: The model takes the reshaped input data `x` (shape: `(n_samples, n_features)`).
- **Predictions**: The model computes the output using the current weights. For a simple dense layer, this is a linear transformation:
  \[
  y = wx + b
  \]

## 2. Loss Calculation
- **Loss Function**: The Mean Absolute Error (MAE) is computed as:
  \[
  \text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
  \]
  where \( y_i \) is the true value and \( \hat{y}_i \) is the predicted value.

## 3. Backward Pass
- **Gradient Calculation**: Gradients of the loss with respect to the model parameters (weights and biases) are calculated usi
- the gradient refers to the slope of the loss (or cost) function with respect to the model parameters (weights and biases)ng backpropagation.

## 4. Weight Update
- **Optimizer Step**: The optimizer (e.g., SGD) updates the model parameters based on the gradients:
  \[
  w := w - \eta \cdot \frac{\partial L}{\partial w}
  \]
  where \( \eta \) is the learning rate.

## 5. Metrics Calculation
- The model computes metrics (e.g., MAE) on the training data to assess performance.

## 6. Epoch Summary
- At the end of each epoch, the average loss and metrics are printed. Ideally, the loss should decrease over the epochs, indicatis learning.

### Example Output


1. Loss Calculation<br>
Loss Function: The loss function (in your case, Mean Absolute Error, or MAE) is used to measure how well the model's predictions match the true values during training. It directly influences the optimization process.
<br>
2. Metric Calculation<br>
Metrics: Metrics are used to evaluate the performance of the model but do not directly affect the training. They provide insights into how well the model is performing on the training or validation data.

```python
tf.random.set_seed(42)
model=tf.keras.Sequential([
    tf.keras.layers.Dense(100,activation=None),
    tf.keras.layers.Dense(1)
])
#model.summary()
model.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.Adam(),metrics=["mae"])
model.fit(x,y,epochs=100)
```

```python
model.predict(np.array([[17.0]]))
```

```python
tf.random.set_seed(42)
model=tf.keras.Sequential([
    tf.keras.layers.Dense(100,activation=None),
    tf.keras.layers.Dense(1)
])
#model.summary()
model.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),metrics=["mae"])
model.fit(x,y,epochs=100)
```

```python
model.predict(np.array([[17.0]]))
```
